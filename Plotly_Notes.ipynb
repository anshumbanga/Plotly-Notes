```python
import pandas as pd
import numpy as np 
import seaborn as sns
import plotly.express as px 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = px.data.iris()
```

```python
df
```

```python
# Scatterplot 



px.scatter(df, x='sepal_length', y='sepal_width', color='species', title='Sepal Length vs Width')
```

Insights:

- We can observe that the setosa species generally has smaller sepal width and length compared to the other two species.
- Versicolor and virginica species have a wider range of sepal lengths and widths.

Purpose:
- Scatter plots are used to visualize the relationship or association between two numerical variables.
 - They help identify patterns, trends, correlations, and outliers in the data.

```python
px.scatter(df, x='sepal_length', y='sepal_width', color='species')
```

# line plot 

Purpose:

- Line plots are commonly used to visualize trends or patterns in data over time or another ordered dimension.
- They are effective for showing changes in a variable's value continuously or sequentially.
(time series analysis)

Components:

- In a line plot, data points are represented by markers (dots), and the markers are connected by lines to show the sequence or progression of the data.
- Each data point typically represents an observation at a specific point in time or order.

Certainly! A bar plot is a type of data visualization that uses rectangular bars to represent the frequency, count, or proportion of categorical data. In Plotly Express (px), you can create bar plots using the px.bar() function. Here's some information about bar plots in Plotly Express:

**Purpose:**

- Bar plots are commonly used to compare the values of categorical variables or to display the distribution of a categorical variable.
- They are effective for visualizing the frequency or proportion of different categories within a dataset.


**Components:**

- In a bar plot, each bar represents a category, and the height (or length) of the bar corresponds to the frequency, count, or proportion of that category.
- The bars are typically arranged along the x-axis, with the category labels displayed on the y-axis.

```python
# barplot

df =sns.load_dataset('tips')
```

```python
df
```

```python
px.bar(df, x='day', y='total_bill')
```

By default, Plotly Express's px.bar() function aggregates data using the sum for numerical variables when plotting bar plots. If you want to display the average instead, you need to perform the aggregation yourself before passing the data to Plotly Express.

```python
# added reset index to make it a dataframe

df_temp = df.groupby('day')['total_bill'].mean().reset_index()
```

```python
df_temp
```

```python
px.bar(df_temp, x='day', y='total_bill')
```

```python
#title

px.bar(df_temp, x='day', y='total_bill',title="Average total bill of the day")
```

```python
# color 

px.bar(df_temp, x='day', y='total_bill',title="Average total bill of the day",color='day')
```

```python
# barmode='group'
# group , overlay , or relative
px.bar(df_temp, x='day', y='total_bill',title="Average total bill of the day",color='day', barmode='group')

# not better representation 
```

```python
px.bar(df, x='day', y='total_bill', color='sex',barmode='group')
```

```python
px.bar(df, x='day', y='total_bill', color='sex',barmode='relative')
```

```python
px.bar(df, x='day', y='total_bill', color='sex',barmode='overlay')
```
